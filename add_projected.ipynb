```python
import os
import shutil
from src.tools import clear_folder

dataset_name = 'dataset-2024-06-12_12-33-29'


mask_folder = f'datasets/{dataset_name}/valid_manually_labelled/mask'
projected_folder = f'datasets/{dataset_name}/valid_manually_labelled/projected'
exps_folder = 'exps'


clear_folder(projected_folder)

def find_and_copy(file_name, source_folder, target_folder):
    for root, _, files in os.walk(source_folder):
        if file_name in files:
            file_path = os.path.join(root, file_name)
            if not 'projected' in file_path:
                continue
            shutil.copy(file_path, target_folder)
            print(f'Copied {file_path} to {target_folder}')
            return True
    return False

for mask_file in os.listdir(mask_folder):
    mask_file_path = os.path.join(mask_folder, mask_file)
    if os.path.isfile(mask_file_path):
        found = find_and_copy(mask_file, exps_folder, projected_folder)
        if not found:
            print(f'File {mask_file} not found in {exps_folder}')
```

```python
import numpy as np
import os
dataset_name = 'dataset-2024-06-12_12-33-29'


mask_folder = f'datasets/{dataset_name}/train/mask_sam'
projected_folder = f'datasets/{dataset_name}/train/mask'

projected_unlabelled = 0
mask_unlabelled = 0
total=0
for file in os.listdir(mask_folder):
    mask = os.path.join(mask_folder, file)
    projected = os.path.join(projected_folder, file)

    # load the two npy array
    mask = np.load(mask)
    projected = np.load(projected)
    
    total += mask.shape[1]*mask.shape[0]
    projected_unlabelled += projected[projected==0].shape[0]
    mask_unlabelled += mask[mask==0].shape[0]
    
reduce = (projected_unlabelled - mask_unlabelled) / projected_unlabelled
print(f'Unlabelled pixels in projected: {projected_unlabelled}')
print(f'Unlabelled pixels in mask: {mask_unlabelled}')
print(f'Unlabelled pixels reduced by {reduce*100:.2f}%')


r2 = 0.751

import numpy as np
rtotal = (projected_unlabelled - mask_unlabelled *(1-r2))/projected_unlabelled
print(f'Unlabelled pixels reduced by {rtotal*100:.2f}%')
       
print("total_label projected_unlabelled", (total - projected_unlabelled*rtotal)/total*100)
print("total_label mask_unlabelled", (total - mask_unlabelled)/total*100)
print("total_label mask_rtotal", (total - projected_unlabelled*(1-rtotal))/total*100)

```

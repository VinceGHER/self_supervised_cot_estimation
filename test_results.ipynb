```python
import pandas as pd
import numpy as np

# Variable names
variable_names = ['DFormer', 'CMX', 'FCNResnet50', 'Asymformer']

# L1Confidence Loss + SAM
mean_t = np.array([0.033193085, 0.029110504, 0.03310102, 0.025285902])
var_t = np.array([0.02460508, 0.01675827, 0.02332902, 0.0138011])

# Unlabeled neg + SAM
mean_unlabeled = np.array([0.037139524, 0.042306036, 0.039716434, 0.039363515])
var_unlabeled = np.array([0.02514956, 0.03494461, 0.0283761, 0.02875417])

# Standard + SAM
mean_standard = np.array([0.1288465, 0.038761906, 0.0408063, 0.04053234])
var_standard = np.array([0.07941166, 0.01778634, 0.03151194, 0.01996617])

# Standard loss
mean_standard_loss = np.array([0.12781858, 0.057330616, 0.044442892, 0.043749318])
var_standard_loss = np.array([0.05728659, 0.01822186, 0.0265831, 0.02066549])


# Calculate improvements
def calculate_improvement(mean_method, mean_standard_loss):
    improvement = np.where(mean_standard_loss != 0, (1 - mean_method / mean_standard_loss) * 100, 0)
    return improvement

improvement_t = calculate_improvement(mean_t, mean_standard_loss)
improvement_unlabeled = calculate_improvement(mean_unlabeled, mean_standard_loss)
improvement_standard = calculate_improvement(mean_standard, mean_standard_loss)

# Create DataFrame
data = {
    'Metric / Method': [
        'Mean (Standard Loss)',
        'Mean (Standard + SAM)',
        'Mean (Unlabeled neg + SAM)',
        'Mean (L1Confidence Loss + SAM)',
    ]
}

# Add each model's data to the DataFrame
for i, name in enumerate(variable_names):
    data[name] = [
        f'{mean_standard_loss[i]:.4f}',
        f'{mean_standard[i]:.4f} ({improvement_standard[i]:.1f}%)',
        f'{mean_unlabeled[i]:.4f} ({improvement_unlabeled[i]:.1f}%)',
        f'{mean_t[i]:.4f} ({improvement_t[i]:.1f}%)',
    ]

df = pd.DataFrame(data)
df.set_index('Metric / Method', inplace=True)

# Display the DataFrame
print(df)
#
```
